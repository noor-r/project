```python
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup

all_matches = []  # Initialize an empty list to store match data

for year in years:
    data = requests.get(standings_url)
    soup = BeautifulSoup(data.text, 'html.parser')  # Specify parser explicitly
    standings_table = soup.select('table.stats_table')[0]
    
    links = [l.get("href") for l in standings_table.find_all('a')]
    links = [l for l in links if '/squads/' in l]
    teams_urls = [f"https://fbref.com{l}" for l in links]
    
    # Get the previous season's standings URL
    try:
        previous_season = soup.select("a.prev")[0].get("href")
        standings_url = f"https://fbref.com{previous_season}"
    except IndexError:
        print(f"No previous season found for year {year}")
        break

    for team_url in teams_urls:
        team_name = team_url.split("/")[-1].replace("Stats", "").replace("-", " ")
        
        # Get match data
        data = requests.get(team_url)
        matches = pd.read_html(data.text, match="Scores & Fixtures")
        
        # Get shooting stats URL
        soup = BeautifulSoup(data.text, 'html.parser')
        links = [l.get("href") for l in soup.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l]
        
        if not links:
            print(f"No shooting data available for team {team_name} in year {year}")
            continue
        
        data = requests.get(f"https://fbref.com{links[0]}")
        shooting = pd.read_html(data.text, match="Shooting")[0]
        shooting.columns = shooting.columns.droplevel()
        
        # Merge data, handling missing columns
        try:
            shooting_subset = shooting[["Date", "Sh", "SoT", "Dist", "PK", "PKatt"]]  # Exclude "FK"
        except KeyError as e:
            print(f"Missing columns in shooting data for team {team_name}: {e}")
            shooting_subset = shooting.loc[:, shooting.columns.intersection(["Date", "Sh", "SoT", "Dist", "PK", "PKatt"])]
        
        try:
            team_data = matches[0].merge(shooting_subset, on="Date")
        except ValueError as e:
            print(f"Error merging data for team {team_name} in year {year}: {e}")
            continue
            
        # Filter for Premier League matches and add metadata
        team_data = team_data[team_data["Comp"] == "Premier League"]
        team_data["Season"] = year
        team_data["Team"] = team_name
        all_matches.append(team_data)
        
        time.sleep(1)  # Rate limiting to avoid server overload
```

```python
match_df = pd.concat(all_matches)
```

```python
match_df.columns = [c.lower() for c in match_df.columns]
```

```python
match_df.to_csv("matches.csv")
```
